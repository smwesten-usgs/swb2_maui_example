# Direct net infiltration

## Introduction

Many significant water budget components may be related to landuse modifications by people. Some of these can act as sources to net infiltration. SWB contains no real mechanisms to simulate the results of most of these landuse modifications, however, gridded estimates of these values may be supplied. These gridded values are added directly to the net infiltration term of the water balance on a daily basis.

SWB version 2.0 contains keywords and definitions for the following sources of direct infiltration:

 - CESSPOOL_LEAKAGE
 - STORM_DRAIN_LEAKAGE 
 - WATER_BODY_LEAKAGE 
 - WATER_MAIN_LEAKAGE 
 - DISPOSAL_WELL_DISCHARGE
 - ANNUAL_DIRECT_NET_INFILTRATION_RATE

All of these direct infiltration sources are supplied to SWB in terms of inches per day, except for the catch-all 'ANNUAL_DIRECT_NET_INFILTRATION_RATE' which is given in inches per year.

These direct infiltration sources may also be supplied in either gridded form or as columns within a lookup table.
